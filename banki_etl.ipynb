## Data Cleaning and Transformation

```python
# drop missing values
banki_df.dropna(subset =['Order_ID', 'Customer_ID'], inplace = True)
banki_df.info()
```

```python
# convert Order_Date from string to DateTime Object
banki_df['Order_Date'] = pd.to_datetime(banki_df['Order_Date'], dayfirst=True)
banki_df.info()
```

```python
# customer table
Customer_df = banki_df[['Customer_ID', 'Customer_Name','Email', 'Phone_Number']].copy().drop_duplicates().reset_index(drop=True)
Customer_df.head()
```

```python
# product table
Product_df = banki_df[['Product_ID','Product_Name', 'Brand', 'Category', 'Price']].copy().drop_duplicates().reset_index(drop=True)
Product_df.head()
```

```python
# shipping address table
Shipping_Address_df = banki_df[['Customer_ID','Shipping_Address', 'City', 'State', 'Country',
       'Postal_Code']].copy().drop_duplicates().reset_index(drop=True)
Shipping_Address_df.index.name = 'Shipping_ID'
Shipping_Address_df = Shipping_Address_df.reset_index()
Shipping_Address_df.head()
```

```python
# Order table
Order_df = banki_df[['Order_ID','Customer_ID','Product_ID','Quantity', 'Total_Price',
       'Order_Date']].copy().drop_duplicates().reset_index(drop=True)
Order_df.head()
```

```python
# payment method table
Payment_Method_df = banki_df[['Order_ID','Payment_Method',
       'Transaction_Status']].copy().drop_duplicates().reset_index(drop=True)
Payment_Method_df.head()
```

```python
# save tables to CSVs
Customer_df.to_csv('dataset/cleaneddata/Customers.csv', index = False)
Product_df.to_csv('dataset/cleaneddata/Products.csv', index = False)
Order_df.to_csv('dataset/cleaneddata/Orders.csv', index = False)
Shipping_Address_df.to_csv('dataset/cleaneddata/Shipping_Address.csv', index = False)
Payment_Method_df.to_csv('dataset/cleaneddata/Payment_Method.csv', index = False)
```

```python

# Create our SQL tables
def create_tables():
    conn = get_db_connection()
    cursor = conn.cursor()
    create_table_query = '''  
                            CREATE SCHEMA IF NOT EXISTS banki;

                            DROP TABLE IF EXISTS banki.Customers CASCADE;
                            DROP TABLE IF EXISTS banki.Products CASCADE;
                            DROP TABLE IF EXISTS banki.Shipping_Address CASCADE;
                            DROP TABLE IF EXISTS banki.Orders CASCADE;
                            DROP TABLE IF EXISTS banki.Payment_Method CASCADE;
                           
                            
                            CREATE TABLE IF NOT EXISTS banki.Customers(
                                Customer_ID UUID PRIMARY KEY,
                                Customer_Name TEXT,
                                Email TEXT,
                                Phone_Number TEXT

                            );


                            CREATE TABLE IF NOT EXISTS banki.Products(
                                Product_ID UUID PRIMARY KEY,
                                Product_Name TEXT,
                                Brand TEXT,
                                Category TEXT,
                                Price FLOAT

                            );


                              CREATE TABLE IF NOT EXISTS banki.Shipping_Address(
                                Shipping_ID SERIAL PRIMARY KEY,
                                Customer_ID UUID,
                                Shipping_Address TEXT,
                                City TEXT,
                                State TEXT,
                                Country TEXT,
                                Postal_Code INTEGER,
                                FOREIGN KEY(Customer_ID) REFERENCES banki.Customers(Customer_ID)

                            );

                              CREATE TABLE IF NOT EXISTS banki.Orders(
                                Order_ID UUID PRIMARY KEY,
                                Customer_ID UUID,
                                Product_ID UUID,
                                Quantity INTEGER,
                                Total_Price FLOAT,
                                Order_Date DATE,
                                FOREIGN KEY(Customer_ID) REFERENCES banki.Customers(Customer_ID),
                                FOREIGN KEY(Product_ID) REFERENCES banki.Products(Product_ID)

                            );

                              CREATE TABLE IF NOT EXISTS banki.Payment_Method(
                                Order_ID UUID ,
                                Payment_Method TEXT,
                                Transaction_Status TEXT,
                                FOREIGN KEY(Order_ID) REFERENCES banki.Orders(Order_ID)

                            ); '''

    cursor.execute(create_table_query)
    conn.commit()
    cursor.close()
    conn.close()
    
```

```python
create_tables()
```

```python
import csv
def load_data_from_csv(csv_path):
    conn = get_db_connection()
    cursor = conn.cursor()
    with open(csv_path, 'r') as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            cursor.execute('''
                            INSERT INTO banki.Customers(Customer_ID, Customer_Name, Email, Phone_Number)
                            VALUES(%s, %s, %s, %s); ''',
                            row

                         )
    conn.commit()
    cursor.close()
    conn.close()

csv_file_path = 'dataset/cleaneddata/Customers.csv'
load_data_from_csv(csv_file_path)
```
